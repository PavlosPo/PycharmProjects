Βασική έννοια στη Pandas είναι το Dataframe. Το Pandas DataFrame είναι ένας δισδιάστατος, μεταβλητού μεγέθους, δυνητικά ετερογενής πίνακας με καθορισμένους άξονες (σειρές και στήλες). 

Το Pandas DataFrame (ως δομή δεδομένων) αποτελείται από τρία βασικά στοιχεία : τα δεδομένα (data), τις σειρές (rows) και τις στήλες (columns).

### Δημιουργία Dataframes

##### Δημιουργία DataFrame καλώντας τον κατασκευαστή
https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.html

```python
df = pd.DataFrame()#κλήση κενού κατασκευαστή, χωρίς ορίσματα
print(df) #κενό DataFrame
```

```python
lst = ['banana', 'orange', 'cherry', 'kiwi']
print(lst)  
```

```python
# Calling DataFrame constructor on list
df = pd.DataFrame(data=lst)#no columns names
print(df)
```

```python
lst = ['banana', 'orange', 'cherry', 'kiwi']
  
# Calling DataFrame constructor on list
df = pd.DataFrame(data=lst, columns = ['fruit'])#added list of column names
print(df)
```

```python
lst = ['banana', 'orange', 'cherry', 'kiwi']
  
# Calling DataFrame constructor on list
df = pd.DataFrame(data=lst, columns = ['fruit'], index = ['fruit1','fruit2','fruit3','fruit4'])#added list of index
print(df)
```

```python
lst = [['banana','yellow'],['orange','orange'], ['cherry','red'],[ 'kiwi','green']]
  
# Calling DataFrame constructor on list
df = pd.DataFrame(data=lst, columns = ['fruit','colour'], index = ['fruit1','fruit2','fruit3','fruit4'])#added list of index
print(df)
```

```python
# Calling DataFrame constructor on list of dicts

data = [{'fruit': 'banana', 'colour': 'yellow'},{'fruit': 'orange', 'colour': 'red', 'price': 2.10}]
df = pd.DataFrame(data, index=['fruit1', 'fruit1'])
print(df) 
```

#### Μερικές χρήσιμες μεθόδοι για γρήγορη εικόνα των δεδομένων
H μέθοδος head() είναι πιο γρήγορος τρόπος να δούμε το n πρώτο αριθμό γραμμών του dataframe

```python
lst = [['banana','yellow',1.10],['orange','orange',0.65], ['cherry','red'],[ 'kiwi','green',2.30]]
df = pd.DataFrame(data=lst, columns = ['fruit','colour','price'], index = ['fruit1','fruit2','fruit3','fruit4'])#added list of index
print(df)
```

```python
print(df.head(2))
```

αντίστοιχα υπάρχει και η μέθοδος tail() που επιστρέψει τον n αριθμό στοιχείων από το τέλος του DataFrame

```python
 print(df.tail())  
```
